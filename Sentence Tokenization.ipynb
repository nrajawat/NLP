```python
import nltk

file1 = "news article.txt" # assign 
file2 = open(file1, "r", encoding ="utf-8")  #open the file
text = file2.read() # read the file and save into text avriable
```

```python
# [1] Detect sentences in the given news article.

sentence_tokens = nltk.sent_tokenize(text)
print ("sentence tokenization is as: \n \n", sentence_tokens)
```

```python
# [2] Tokenize each sentence into words

word_tokens = nltk.word_tokenize(text)
print (" Tokenized file: \n\n", word_tokens)
```

```python
# [3] Perform Part-of-Speech (POS) on each sentence
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

stoppers = set(stopwords.words('english'))
tokenized_file = sent_tokenize(text)
for i in tokenized_file:
     
    # Word tokenizers is for finding words and punctuation in a string
    List_of_words = nltk.word_tokenize(i)
 
    # removing stop words from wordList
    List_of_words = [w for w in List_of_words if not w in stoppers]
 
    #  Using a Tagger. Which is part-of-speech
    # tagger or POS-tagger.
    tagged_file = nltk.pos_tag(List_of_words)
 
    print(tagged_file)
```

```python
# [4] Find name entities including person’s name entities and locations 
import spacy
from spacy import displacy

NLP = spacy.load("en_core_web_sm")
NLP.pipe_names

doc = NLP(text)

for ent in doc.ents:
    print(ent.text,"->", ent.label_ , "|", spacy.explain(ent.label_))
    
```
